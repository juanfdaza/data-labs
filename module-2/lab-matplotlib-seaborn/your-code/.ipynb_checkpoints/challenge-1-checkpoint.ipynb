# Challenge 1

#### Import all the libraries that are necessary.

```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,3,1)
plt.plot(x,y,lw=3,color="b", ls='--');
plt.subplot(1,3,2)
plt.plot(x,z,lw=3, color="r",ls='-');
plt.subplot(1,3,3)
plt.plot(x,y,lw=3,color="b", ls='--');
plt.subplot(1,3,3)
plt.plot(x,z,lw=3, color="r",ls='-');
```

```python
# 2nd method
fig, ax = plt.subplots(1,3)
ax[0].plot(x,y,color="b", lw=3, ls='--');
ax[1].plot(x,z,color="y", lw=3, ls='-');
ax[2].plot(x,y,color="b", lw=3, ls='--');
ax[2].plot(x,z,color="y", lw=3, ls='-');
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, ax = plt.subplots(1,3,figsize=(8,8))
ax[0].plot(x,y,color="b", lw=3, ls='--');
ax[1].plot(x,z,color="y", lw=3, ls='-');
ax[2].plot(x,y,color="b", lw=3, ls='--');
ax[2].plot(x,z,color="y", lw=3, ls='-');
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, ax = plt.subplots(1,3,figsize=(8,8))
ax[0].plot(x,y,color="b", lw=3, ls='--');
ax[0].set_xlabel('x values');
ax[0].set_ylabel('y values');
ax[1].plot(x,z,color="y", lw=3, ls='-');
ax[1].set_xlabel('x values');
ax[1].set_ylabel('y values');
ax[2].plot(x,y,color="b", lw=3, ls='--');
ax[2].plot(x,z,color="y", lw=3, ls='-');
ax[2].set_xlabel('x values');
ax[2].set_ylabel('y values');
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig, ax = plt.subplots(1,2,figsize=(8,8))

# First we define the y function
y = x**2
z = np.exp(x)
ax[0].plot(x,y,color="b", lw=3,);
ax[0].set_xlabel('x values');
ax[0].set_ylabel('y values');
ax[0].plot(x,z,color="r", lw=2, ls='--');

## Now log scale

ax[1].plot(x,y,color="b", lw=3,);
ax[1].set_xlabel('x values');
ax[1].set_ylabel('y values');
ax[1].plot(x,z,color="r", lw=2, ls='--');
ax[1].set_yscale("log")
```
